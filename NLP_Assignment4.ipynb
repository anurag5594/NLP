```python
import spacy
from spacy.pipeline import EntityRuler
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

```

```python
nlp = spacy.blank("en")
```

```python
ruler = nlp.add_pipe("entity_ruler", before="ner" if "ner" in nlp.pipe_names else None)

patterns = [
    {"label": "PERSON", "pattern": [{"LOWER": "elon"}, {"LOWER": "musk"}]},
    {"label": "ORG", "pattern": [{"LOWER": "tesla"}]},
    {"label": "ORG", "pattern": [{"LOWER": "spacex"}]},
    {"label": "GPE", "pattern": [{"LOWER": "united"}, {"LOWER": "states"}]}
]

ruler.add_patterns(patterns)
```

```python
text = "Elon Musk is the CEO of Tesla and SpaceX based in the United States."
doc = nlp(text)

```

```python
predicted_entities = [(ent.text, ent.label_) for ent in doc.ents]
print(predicted_entities)
```

```python
true_entities = [
    ('Elon Musk', 'PERSON'),
    ('Tesla', 'ORG'),
    ('SpaceX', 'ORG'),
    ('United States', 'GPE')
]
```

```python
y_true = [label for _, label in true_entities]
y_pred = [label for _, label in predicted_entities]
```

```python
precision = precision_score(y_true, y_pred, average='macro')
recall = recall_score(y_true, y_pred, average='macro')
f1 = f1_score(y_true, y_pred, average='macro')
accuracy = accuracy_score(y_true, y_pred)

print("Accuracy :", accuracy)
print("Precision:", precision)
print("Recall   :", recall)
print("F1-score :", f1)
```

```python
import nltk
from nltk.tokenize import (
    WhitespaceTokenizer,
    wordpunct_tokenize,
    TreebankWordTokenizer,
    TweetTokenizer,
    MWETokenizer
)
from nltk.stem import PorterStemmer, SnowballStemmer
from nltk.stem import WordNetLemmatizer
```

```python
text = "NLTK is amazing! I'm learning NLP, and it's fun ðŸ˜Š"
```

```python
wt = WhitespaceTokenizer()
print(wt.tokenize(text))
```

```python
tbt = TreebankWordTokenizer()
print(tbt.tokenize(text))
```

```python
tweet_tokenizer = TweetTokenizer()
print(tweet_tokenizer.tokenize(text))
```

```python
lemmatizer = WordNetLemmatizer()

words = ["running", "better", "cars", "went"]

print("Default:")
for w in words:
    print(w, "â†’", lemmatizer.lemmatize(w))
    
print("\nWith POS tags:")
print("running â†’", lemmatizer.lemmatize("running", pos="v"))
print("better â†’", lemmatizer.lemmatize("better", pos="a"))
print("went â†’", lemmatizer.lemmatize("went", pos="v"))
```

```python
documents = [
    "I learn natural language processing",
    "Natural language processing is fun",
    "I learn learning NLP"
]
```

```python
count_vectorizer = CountVectorizer()
bow_matrix = count_vectorizer.fit_transform(documents)

print("Vocabulary:")
print(count_vectorizer.get_feature_names_out())

print("\nCount Occurrence Matrix:")
print(bow_matrix.toarray())
```

```python
count_vectorizer_norm = CountVectorizer(norm='l2')
bow_norm = count_vectorizer_norm.fit_transform(documents)

print("\nNormalized Count Matrix:")
print(bow_norm.toarray())
```
